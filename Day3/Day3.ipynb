```python
import numpy as np
import pandas as pd
import string
```

# Task 1

## Raw input

```python
# Using readlines()
input_txt = open('/Users/michael/Documents/GitHub/AoC_2022/Day3/input.txt', 'r')
Lines = input_txt.readlines()
df = pd.DataFrame(columns=["Pack","Items"])

count = 0
# Strips the newline character
for i,line in enumerate(Lines):
    count += 1
    df.loc[i] = ([count, line.strip()])
    
# Creating lists and columns for pack compartments
comp1 = []
comp2 = []

for x in df['Items']:
    half = int(len(x)/2)
    first = x[0:(half)]
    second = x[-half:]
    comp1.append(first)
    comp2.append(second)
    
df["Comp1"] = comp1
df["Comp2"] = comp2


df.head()
```

## Variables

```python
# Creating a dict with upper and lower case letters and values for mapping
lower_case = [x for x in string.ascii_lowercase]
upper_case = [x for x in string.ascii_uppercase]

priority_dict = {}

for i,x in enumerate(lower_case):
    priority_dict[x] = i+1
    
for i,x in enumerate(upper_case):
    priority_dict[x] = i+27
```

## Analysis

```python
# Comparing both compartments of each pack to get the common item
matching = []
for i,x in enumerate(df['Comp1']):
    a=list(set(x)&set(df['Comp2'][i]))
    for i in a:
        matching.append(i)

df["Matching items"] = matching
df.head()
```

```python
# Mapping the priority values

df["Matching value"] = df["Matching items"].map(priority_dict)
df.head()
```

```python
df['Matching value'].sum()
```

# Part 2

## Variables

```python
elves = len(df)
group_count = int(elves/3)
group_range = [x for x in range(1,group_count+1)]
```

## Aanalysis

```python
# Creating a column that holds the group number for each elf
group_list = []
for x in group_range:
    group_list = group_list + [x]*3 
    
df["Group"] = group_list
df.head()
```

```python
# Creating a column to hold matching items for each group and a list to hold each group's identifier
group_identifier = []
matching_list = []

for x in group_range:
    df_group = df[df["Group"]==x]
    first = list(df_group["Items"])[0]
    second = list(df_group["Items"])[1]
    third = list(df_group["Items"])[2]
    a=list(set(first)&set(second)&set(third))
    for i in a:
        group_identifier.append(i)
        matching_list = matching_list+([i]*3)

df['Group match'] = matching_list
df.head()
```

```python
# Using group identifier to create a new data frame
group_df = pd.DataFrame(group_identifier, columns=['Match'])
group_df.head()
```

```python
group_df['Value'] = group_df['Match'].map(priority_dict)
group_df.head()
```
